
### Test Cases

#### Test Case 1: Admin User Creation
- **Test Case ID**: TC101
- **Test Description**: Verify that an administrator can create a new user account.
- **Preconditions**: Admin must be logged in.
- **Test Steps**:
  1. Navigate to the Admin dashboard.
  2. Go to the ‘Create User’ section.
  3. Enter valid user details (name, email, role).
  4. Click on the 'Create User' button.
- **Expected Result**: A new user account is created, and a confirmation message is displayed.
- **Status**: To be filled during execution.

#### Test Case 2: Password Reset Functionality
- **Test Case ID**: TC102
- **Test Description**: Ensure that the password reset function sends an email to the user’s registered email address.
- **Preconditions**: User has forgotten their password.
- **Test Steps**:
  1. Go to the login page.
  2. Click on the ‘Forgot Password’ link.
  3. Enter the registered email address.
  4. Submit the request.
- **Expected Result**: The system sends a reset password link to the provided email.
- **Status**: To be filled during execution.
